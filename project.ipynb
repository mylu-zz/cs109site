# Give It Five Stars: A Data Analysis of Amazon Product Reviews

The practical motivation for our project comes from the rise of online commerce as the main method of acquiring goods and services. Amazon provides convenience and competitive prices, acting as a realistic alternative to traditional stores. Furthermore, Amazon offers an ever growing array of services from basic online shopping to web server hosting. Previous studies have shown that consumer reviews have an impact on sales, especially for niche products like video games (Zhu and Feng, 2010). For example, a product with a 5-star average tends to have better sales than a similar product with a lower rating.

Thus, in predicting the ratings from the text of these reviews, we want to determine whether the number of stars for a review is consistent with the corresponding textual sentiment. Is there inflation or deflation in ratings for certain products? Are the reviews independent and reliable in terms of the language used in them and the ratings given, or do they follow any trend based on date posted, a “follow-the-leader” type mentality? Are there differences in the ability to predict rating depending on the category of product being evaluated?

In being able to summarize the text review with a star rating, we can help people avoid reading overly long reviews and also cancel out any subjectivity that might be involved in giving a star rating.

Reports of partiality in amazon reviews have also been publicized This is especially true with book reviews, where it was discovered that some authors wrote reviews for their own books and wrote negative reviews for rival books (The Guardian). This leads to the question of what the distribution is of negative vs. positive reviews, and highlights the importance of exploring consumer reviews for biases and linkages. This information could help consumers reevaluate how much weight to put on consumer reviews when shopping on amazon. Furthermore, Amazon is the largest source of consumer reviews, and many consumers compare products that they plan to buy elsewhere with reviews on Amazon, so our findings could have applicability to online shopping in general (Business Wire).

```python
def parse(filename):
  f = gzip.open(filename, 'r')
  entry = {}
  for l in f:
    l = l.strip()
    colonPos = l.find(':')
    if colonPos == -1:
      yield entry
      entry = {}
      continue
    eName = l[:colonPos]
    rest = l[colonPos+2:]
    entry[eName] = rest
  yield entry

def word_feats(words):
    return dict([(words, True)])

arts_fname = 'Arts.txt.gz'
```

# Bag of Words and n-gram models

With a bag of words and n-gram approach we want to see whether the frequency of words in the text of a review can serve as a predictor for review score. By breaking up the text of the review into n-grams, basically splitting the text based on spaces and period we can create a list of tokens, e.g. "I", "he", "awesome", "!". We create from this list of tokens a frequency distribution that tallies the number of occurences for each token in a dictionary. This data can be passed to a classifier to be trained on, and hopefully classify a review text to a score. 

A simple bag of words/1-gram/unigram model breaks up a text into single tokens or words. A bigram/2-gram model breaks text up into groups of two words, e.g. "I am", "am a", "a human". A trigram/3-gram model breaks up text into groups of three words, e.g. "I am a", "am a human".

```python
#get tokenizations for a list of reviews
#add in removing stop words here
from nltk import tokenize
def get_tokens(review_list):
    return [tokenize.word_tokenize(r.lower()) for r in review_list]
# arts_reviews_tokens = get_tokens(arts_reviews)
```

```python
#create unigram frequency distributions for each list of tokens
def get_unigram_fds(tokens):
    return [probability.FreqDist(t) for t in tokens]
# arts_reviews_unigram_fds = get_unigram_fds(arts_reviews_tokens)
```

```python
#create bigrams frequency distributions for each list of tokens
from nltk import collocations
from nltk.probability import ELEProbDist, FreqDist
from nltk import NaiveBayesClassifier
from collections import defaultdict
from nltk import metrics
from nltk import bigrams

def get_bigram_fds(tokens):
    return [probability.FreqDist(bigrams(t)) for t in tokens]
# arts_reviews_bigram_fds = get_bigram_fds(arts_reviews_tokens)
```

```python
#create trigrams frequency distributions for each list of tokens
from nltk import trigrams
def get_trigram_fds(tokens):
    return [probability.FreqDist(trigrams(t)) for t in tokens]
# arts_reviews_trigram_fds = get_trigram_fds(arts_reviews_tokens)
```

```python
def get_unibigram_fds(tokens):
    return [probability.FreqDist(t) + probability.FreqDist(bigrams(t)) for t in tokens]
# arts_reviews_unibigram_fds = get_unibigram_fds(arts_reviews_tokens)
```

```python
def get_allgrams_fds(tokens):
    return [probability.FreqDist(t) + probability.FreqDist(bigrams(t)) + probability.FreqDist(trigrams(t)) for t in tokens]
# arts_reviews_allgram_fds = get_allgrams_fds(arts_reviews_tokens)
```

```python
#zip together tokens and labels
def zip_token_label(tokens, labels):
    return zip(tokens, labels)

# arts_train = zip_token_label(arts_x_train, arts_y_train)
# arts_test = zip_token_label(arts_x_test, arts_y_test)
```

```python
#unzip tokens and labels
def unzip_token_label(zipped):
    tokens=[]
    labels=[]
    for t, l in zipped:
        tokens.append(t)
        labels.append(l)
    return tokens, labels
# unzip_token_label(arts_test)
```

```python
#sort into five separate lists: reviews of score 1, 2, 3, 4, 5
#takes in zipped list
def sort_by_score(zipped):
    df = pd.DataFrame(zipped)
    df.columns = ['tokens', 'score']
    lst = []
    for i in range(5):
        df_i = df[df['score'] == (str(i+1)+'.0')]
        lst.append(zip_token_label(list(df_i['tokens']), list(df_i['score'])))
    return lst
# arts_train_sorted = sort_by_score(arts_train)
```

```python
#limit training lists down to length of shortest lists\

#find length of shortest list
def shortest_len(train):
    lens = []
    for l in train:
        lens.append(len(l))
    return min(lens)

#take in training lists sorted by score and return lists of the same length, based on length of shorted list
def limit_training(train):
    shortest = shortest_len(train)
    return [l[:shortest] for l in train]

# arts_train_limited = limit_training(arts_train_sorted)
# print len(arts_train_limited[3])
```

Below is the function to run our n-grams models. We have created five that incorporate different n-gram information: unigrams, bigrams, trigrams, unigrams and bigrams (called unibigrams) and all three n-gram models (allgrams). For each product category that is examined we limited the dataset to reviews since 2004 because a range of 10 years provided enough data while limiting outdated reviews. 

We split product reviews into training and test sets. To deal with class imbalance in the training sets we limited the set to x reviews for each score, where x in the minimum of the number of reviews for each score. Without this adjustment to the training set, the classifiers would all guess 5.0 because doing so achieved high accuracy as 5.0 is the most common score across categories. 

For the classifier we used tf-idf weighting which assigns a weight to term t in document d. This weight is highest when t occurs many times in a small number of documents, lower when it occurs few times or in many documents, and lowest when it occurs in almost all documents. 
We use a chi-squared statistic to select the most informative features and use a multinomial bayes classifier because there are five classes (five scores). 

```python
# package the bag of words and n-grams classifier into a function
# takes in a file name and returns [[list of trained classifiers], [accuracies for test sets]], order for each sublist is [uni, bi, tri, unibi, all]

def run_ngrams(fname):
    df = create_df(fname)
    df = limit_years(df, 2004)
    reviews = get_reviews(df)
    scores = get_scores(df)
    reviews_tokens = get_tokens(reviews)
    x_train, x_test, y_train, y_test = create_test_train(reviews_tokens, scores, 0.33)
    
    train = zip_token_label(x_train, y_train)
    test = zip_token_label(x_test, y_test)

    train_sorted = sort_by_score(train)
    train_limited = limit_training(train_sorted)
    train_limited_conc = conc_lists(train_limited)
    train_unigram = convert_to_ngram(train_limited_conc, 1)
    train_bigram = convert_to_ngram(train_limited_conc, 2)
    train_trigram = convert_to_ngram(train_limited_conc, 3)
    train_unibigram = convert_to_ngram(train_limited_conc, 12)
    train_allgram = convert_to_ngram(train_limited_conc, 123)
    test_unigram = convert_to_ngram(test, 1)
    test_bigram = convert_to_ngram(test, 2)
    test_trigram = convert_to_ngram(test, 3)
    test_unibigram = convert_to_ngram(test, 12)
    test_allgram = convert_to_ngram(test, 123)

    pipeline1 = Pipeline([('tfidf', TfidfTransformer()),
                     ('chi2', SelectKBest(chi2, k=100)),
                     ('nb', MultinomialNB())])
    pipeline2 = Pipeline([('tfidf', TfidfTransformer()),
                     ('chi2', SelectKBest(chi2, k=100)),
                     ('nb', MultinomialNB())])
    pipeline3 = Pipeline([('tfidf', TfidfTransformer()),
                     ('chi2', SelectKBest(chi2, k=100)),
                     ('nb', MultinomialNB())])
    pipeline4 = Pipeline([('tfidf', TfidfTransformer()),
                     ('chi2', SelectKBest(chi2, k=100)),
                     ('nb', MultinomialNB())])
    pipeline5 = Pipeline([('tfidf', TfidfTransformer()),
                     ('chi2', SelectKBest(chi2, k=100)),
                     ('nb', MultinomialNB())])

    classif = SklearnClassifier(pipeline1)
    bg_classif = SklearnClassifier(pipeline2)
    tg_classif = SklearnClassifier(pipeline3)
    ubg_classif = SklearnClassifier(pipeline4)
    ag_classif = SklearnClassifier(pipeline5)
    
    classif.train(train_unigram)
    bg_classif.train(train_bigram)
    tg_classif.train(train_trigram)
    ubg_classif.train(train_unibigram)
    ag_classif.train(train_allgram)

    acc_uni = nltk.classify.accuracy(classif, test_unigram)
    acc_bi= nltk.classify.accuracy(bg_classif, test_bigram)
    acc_tri = nltk.classify.accuracy(tg_classif, test_trigram)
    acc_unibi = nltk.classify.accuracy(ubg_classif, test_unibigram)
    acc_all = nltk.classify.accuracy(ag_classif, test_allgram)
    
    return [[classif, bg_classif, tg_classif, ubg_classif, ag_classif], [acc_uni, acc_bi, acc_tri, acc_unibi, acc_all]]

arts_results = run_ngrams(arts_fname)
print arts_results[1]
    
```

# Sentiment Analysis

With sentiment analysis, we try to determine the polarity of the text (i.e. positive or negative sentiment).  Using the polarity of the review text, we then predict the review rating.  Reviews with many positive words (i.e. good, excellent, wonderful) should have high ratings while reviews with many negative words (i.e. terrible, wasteful, worst) should have low ratings.

To begin with, we assume that adjectives found in a small subset of 5.0 star reviews are positive and adjectives found in a small subset of 1.0 star reviews are negative.  We train a naive Bayes classifier based on this data so that it may classify adjectives into two categories: positive or negative.

## Training classifier for sentiment analysis

We split the text of each review for 5 star and 1 star ratings into words and only consider adjectives.  We train the naive Bayes classifier based on these adjectives.  We then print the most informative features.

```python
import gzip
import json as simplejson #import simplejson
import numpy as np
from sklearn.lda import LDA
import collections
import math

import nltk
from nltk.classify import NaiveBayesClassifier

def train_sentiment_analysis(fname):
  products = []
  for product in parse(fname):
    products.append(product)

  # gets a list of positive adjectives (defined now as adjs in reviews with 4 or 5 stars)
  pos = []
  neg = []
  pos_reviews = 0
  neg_reviews = 0
  NUM_REVIEWS = 500

  for p in products:
    if 'review/score' in p.keys():
      if p['review/score'] == '5.0' and pos_reviews <= NUM_REVIEWS: # or p['review/score'] == '4.0':
        pos_reviews = pos_reviews + 1

        # make string of the review
        review = ''
        for word in p['review/text']:
          review = review + word
        text = nltk.word_tokenize(review)
        tags = nltk.pos_tag(text)
    
        for tag in tags:
          if tag[1] == 'JJ':
      	    pos.append(tag[0].lower())

      elif p['review/score'] == '1.0' and neg_reviews <= NUM_REVIEWS: # or p['review/score'] == '4.0':
        neg_reviews = neg_reviews + 1

        # make string of the review
        review = ''
        for word in p['review/text']:
          review = review + word
        text = nltk.word_tokenize(review)
        tags = nltk.pos_tag(text)
    
        for tag in tags:
          if tag[1] == 'JJ':
      	    neg.append(tag[0].lower())

  pos_features = [(word_feats(f), 'positive') for f in pos ]
  neg_features = [(word_feats(f), 'negative') for f in neg ]

  trainfeats = neg_features + pos_features
 
  classifier = NaiveBayesClassifier.train(trainfeats)
  classifier.show_most_informative_features()
  return classifier, products
  # print 'accuracy:', nltk.classify.util.accuracy(classifier, testfeats)
```

## Use sentiment analysis to predict rating

For each review for which we are trying to predict the rating, we first split the text of the review into words and only consider the adjectives.  We then classify these adjectives as positive or negative.  Using these classifications, we make a prediction (i.e. adding one to the score if positive adjective encountered).

If there are no positive or negative adjectives, then we assume that the score is 3.0 stars as the text is effectively neutral.  No positive adjectives creates a default rating of 1.0 star.

We also find the average error by summing the squared difference between actual and predicted ratings, dividing by the number of review ratings predicted, and then taking the square root.

```python
def run_sentiment_analysis(classifier, products):
  # use classifier to predict ratings
  predictions = []
  actual = []
  NUM_REVIEWS = 1000
  reviews = 0

  for p in products:
    if 'review/score' in p.keys() and reviews <= NUM_REVIEWS:
      score = 0.0
      review = ''
      reviews = reviews + 1

      actual.append(p['review/score'])

      for word in p['review/text']:
        review = review + word
      text = nltk.word_tokenize(review)
      tags = nltk.pos_tag(text)

      num_tags = 0
    
      for tag in tags:
        if tag[1] == 'JJ':
          num_tags = num_tags + 1
          if classifier.classify(word_feats(tag[0].lower())) == 'positive':
            score = score + 1.0

      # if no relevant adjectives, then label review as neutral
      if num_tags == 0:
        score = 3.0
      else:
        score = score / num_tags
        score = math.ceil(score * 5) + 0.0
        if score == 0:
          score = 1.0
      predictions.append(score)

  error = [math.pow(float(a) - b, 2) for a, b in zip(actual, predictions)]
  error = sum(error)
  error = error / NUM_REVIEWS
  error = math.sqrt(error)

  print 'average error per review: ' + str(error)
  return predictions, actual
```

# Latent Dirichlet Allocation

In our third approach, we use latent dirichlet allocation (LDA) to model product reviews as a mixture of hidden topics. Each document is expressed as a vector of topic weights, and each word in the document is generated by a certain topic, which has a multinomial probability distribution over words. LDA assumes a generative model of document formation, where we generate a document by choosing the number of words N from a Poisson distribution and choosing a prior topic mixture according to a dirichlet distribution. Then, we generate words in the document by randomly choosing a topic according to a multinomial distribution (with weights set from the dirichlet prior) and then using that topic to generate words according to a multinomial distribution. 

LDA reverses this process to try to discover these topics using collapsed Gibbs sampling. In this method, we go through each document in our corpus and preliminarily assign a word in the document to a topic. This initializes our topic distributions for each document and word distributions for each topic. We try to update these values by iterating through each topic t for each word w in each document d and update two values. We first find P(t | d), which is the proportion of words assigned to topic t in d, and P(w | t), the proportion of the number of times word w is assigned to topic t over the number of total assignments to topic t. We reassign the word w to a new topic t' with probability p(t' | d)p(w | t'). After many iterations, we expect these values to converge, and thus we can find the mixture of topics from the proportion of words in each document assigned to each topic. We can also calculate the words most commonly generated from a topic from the proportion of words assigned to each topic.

However, collapsed Gibbs Sampling does not assume independence between documents or topics, so we cannot efficiently parallelize this process. For that reason, we only look at a subset of 10000 reviews from each dataset. We perform some preliminary steps to clean the data, removing special characters and using NLTK's snowball stemmer.

```python
products = all_set
targets = []
titles = []
words = set()
stemmer = SnowballStemmer("english")
tdm = textmining.TermDocumentMatrix()

def clean_word(s):
  return word.strip('\\/?!*()"\'')   

for product in np.random.choice(products,10000,replace=False):
  try:
    targets.append(int(float(product['review/score'])))
    review_text = product['review/text']
    review_text = review_text.replace('.',' ')
    review_text = review_text.replace(',',' ')
    review_text = review_text.replace(';',' ')
    review_text = review_text.replace(':',' ')
    review_text = review_text.replace('-',' ')
    review_text = review_text.replace('#',' ')
    review_text = review_text.replace('&',' ')
    review_text = review_text.replace('!',' ')
    tdm.add_doc(' '.join([stemmer.stem(clean_word(word)) for word in review_text.split()]))
    for word in review_text.split():
      words.add(stemmer.stem(clean_word(word)))
  except:
    pass

rows = tdm.rows()
X = np.zeros((len(targets),len(rows.next())))
j = 0
for row in rows:
  X[j,:] = list(row)
  j += 1
```

# Time correlation analysis

In our exploratory data analysis we want to determine whether early reviews have a follow-the-leader type effect on later reviews. Thus we decided to look at a moving average of review scores for products that were reviewed many times. This analysis would show us whether review averages changed significantly over time or whether trends set by early reviews were highly influential. This information is important in determining whether or not consumers should trust reviews to be independent of earlier reviews. 

```python
#returns a dictionary
import itertools  
def create_dict(fname):
    products = []
    for product in parse(fname):
        if 'review/score' in product.keys():
            products.append(product)
    newdict = collections.defaultdict(list)
    for p in products:
        newdict[p['product/productId']] += [(p['review/score'], p['review/time'], p['product/price'])]
    return newdict

#limit the dictionary to products that have greater than n reviews

def limit_dict(d, n):
    newdict = collections.defaultdict(list)
    for p in d.keys():
        if len(d[p]) >=n:
            d[p].sort(key = lambda tup: float(tup[1]))
            newdict[p] = d[p]
    return newdict
    
def take(n, iterable):
    return list(itertools.islice(iterable, n))

# art_dict = limit_dict(create_dict(arts_fname), 200)

# print take(6, art_dict.iteritems())[0][1]
    
```

```python
#unzips a list of (score, time, price) tuples and returns the scores and times
def unzip_score_time_price(lst):
    scores = []
    times = []
    prices = []
    for x in lst:
        score, time, price = x
        scores.append(float(score))
        times.append(float(time))
#         prices.append(float(price))
    return (scores, times)
```

```python
# calculates the moving average for a list of values based on a window of values
def moving_average(values, window):
    wts = np.repeat(1.0, window)/float(window)
    ma = np.convolve(values, wts, 'valid')
    return ma

```

**Reasoning** There are several parameters that we specified to avoid being overwhelmed by the data. First, we look at products from each category that had 200 reviews or more. This would provide us with products that have likely been reviewed over a long period of time so that we can see trends and fluctuations more clearly. 
We chose randomly six products from each category to give us a representative look at the moving average of reviews.
Finally, we chose a window of the past 20 reviews on which to base our moving average. Most people will look at several reviews before writing their own so we thought this to be a reasonable number so that a larger trend could be seen, but the effects of individual reviews were not insignificant. 
The overall average of the review scores is shown by a horizontal line to judge whether the moving averages may converge to the true average.

```python
# run the moving average analysis on a given category
def run_ma_analysis(fname):
    category_name = fname.split('.')[0]
    num_reviews = 200
    num_products = 6
    window = 20
    product_dict = limit_dict(create_dict(fname), num_reviews)
    to_run = take(num_products, product_dict.iteritems())
    fig = plt.figure(figsize=(14,4))
    i = 1
    j = 1
    for p in to_run:
        if i > 2:
            fig.savefig('', )
            plt.show()
            fig = plt.figure(figsize=(14,4))
            i = 1
        x = fig.add_subplot(1,2,i)
        i+= 1
        pid = p[0]
        data = p[1]
        scores, times = unzip_score_time_price(data)
        overall_avg = np.mean(scores)
        times = pd.to_datetime(times, unit = 's')
        x.scatter(times[(window-1):], moving_average(scores, window))
        x.axhline(y = overall_avg)
        x.set_title('Moving Average analysis for ' + pid + ' in ' + category_name)
        x.set_ylabel('Average for last ' + str(window) + ' reviews')
        x.set_xlabel('Date')
    plt.show()
        
run_ma_analysis(arts_fname)
    
    
    
```
